```python
def find_table_entry(dir_path, file, attr, attr_value):
    df = pd.read_csv(f"{dir_path}/{file}.csv")
    try:
        res = df.loc[df[attr] == attr_value]
    except:
        print("Attribute Mismatch")
    
    return res
```

```python
foodb_ordered_data = {}
elements_encounterd = set()
c = 0


for index, row in row_data:
    if c > 10000:
        break
    c += 1
    elem = {}

    # Attributes with repeated rows
    if row['id_x'] in elements_encounterd:
        # This could be empty
        if type(row['orig_food_common_name']) == str or type(row['orig_source_name']) == str or type(row['orig_content']) == str:
            foodb_ordered_data[row['id_x']]['orig_food'].append({
                'orig_food_common_name': row['orig_food_common_name'],
                'orig_source_name': row['orig_source_name'],
                'orig_content': f"{row['orig_content']} {row['orig_unit']}"
                })

        if row['source_type'] == 'Nutrient':
            nutr_row = find_table_entry(dir_path, 'Nutrient', 'id', row['source_id'])
            if nutr_row['name'].values.size != 0:
                foodb_ordered_data[row['id_x']]['Nutrient'].append(nutr_row['name'].values[0])
        else:
            cmp_row = find_table_entry(dir_path, 'Compound', 'id', row['source_id'])
            health_effect_exists = True
            try:
                health_effect_id = find_table_entry(dir_path, 'CompoundsHealthEffect', 'compound_id', row['source_id'])['health_effect_id'].values[0]
                health_effect = find_table_entry(dir_path, 'HealthEffect', 'id', health_effect_id)
            except:
                health_effect_exists = False
            if cmp_row['name'].values.size != 0:
                foodb_ordered_data[row['id_x']]['Compound'].append({
                    'name': cmp_row['name'].values[0],
                    'description': cmp_row['annotation_quality'].values[0],
                    'kingdom': cmp_row['kingdom'].values[0],
                    'superklass': cmp_row['superklass'].values[0],
                    'klass': cmp_row['klass'].values[0],
                    'subklass': cmp_row['subklass'].values[0],
                    'health_effect_name': '' if not health_effect_exists else health_effect['name'].values[0],
                    'health_effect_desc': '' if not health_effect_exists else (f"{'' if type(health_effect['chebi_name'].values[0]) != str else health_effect['chebi_name'].values[0]}{'' if type(health_effect['description'].values[0]) != str else '. ' + health_effect['description'].values[0]}")
                })
    else:
        elements_encounterd.add(row['id_x'])
        elem['name'] = row['name']
        elem['name_scientific'] = row['name_scientific']
        elem['description'] = row['description']
        elem['wikipedia_id'] = row['wikipedia_id']
        elem['food_group'] = row['food_group']
        elem['food_subgroup'] = row['food_subgroup']
        elem['orig_food'] = []
        if type(row['orig_food_common_name']) == str or type(row['orig_source_name']) == str or type(row['orig_content']) == str:
            elem['orig_food'].append({
                'orig_food_common_name': row['orig_food_common_name'],
                'orig_source_name': row['orig_source_name'],
                'orig_content': f"{row['orig_content']} {row['orig_unit']}"
                })
        elem['Nutrient'] = []
        elem['Compound'] = []
        if row['source_type'] == 'Nutrient':
            nutr_row = find_table_entry(dir_path, 'Nutrient', 'id', row['source_id'])
            if nutr_row['name'].values.size != 0:
                elem['Nutrient'].append(nutr_row['name'].values[0])
        else:
            cmp_row = find_table_entry(dir_path, 'Compound', 'id', row['source_id'])
            health_effect_exists = True
            
            try:
                health_effect_id = find_table_entry(dir_path, 'CompoundsHealthEffect', 'compound_id', row['source_id'])['health_effect_id'].values[0]
                health_effect = find_table_entry(dir_path, 'HealthEffect', 'id', health_effect_id)
            except:
                health_effect_exists = False
            
            
            if cmp_row['name'].values.size != 0:
                elem['Compound'].append({
                    'name': cmp_row['name'].values[0],
                    'description': cmp_row['annotation_quality'].values[0],
                    'kingdom': cmp_row['kingdom'].values[0],
                    'superklass': cmp_row['superklass'].values[0],
                    'klass': cmp_row['klass'].values[0],
                    'subklass': cmp_row['subklass'].values[0],
                    'health_effect_name': '' if not health_effect_exists else health_effect['name'].values[0],
                    'health_effect_desc': '' if not health_effect_exists else (f"{'' if type(health_effect['chebi_name'].values[0]) != str else health_effect['chebi_name'].values[0]}{'' if type(health_effect['description'].values[0]) != str else '. ' + health_effect['description'].values[0]}")
                })
        ## ADD PATHWAY IN HEALTH EFFECTS
        foodb_ordered_data[row['id_x']] = elem
```
