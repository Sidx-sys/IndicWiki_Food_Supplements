```python
import json
```

```python
with open("finetuned_hindi_data.json", 'r') as f:
    data = json.load(f)
```

### Need to understand and create Infobox

```python
from jinja2 import Template
```

```python
data['1']
```

```python
content="""=={{food['name']}}==
{{food['description']}} खाद्य समूह {{food['food_group']}} हैं, विशेष रूप से {{food['food_subgroup']}} के उपसमूह के अंतर्गत आता है।

{% raw %}{{Infobox {% endraw %} food
| name = {{food['name']}}
|alternate_name = (वैज्ञानिक नाम) {{food[name_scientific]}}
| serving size = 100 g
{% if 'energy' in food['nutritional_info'] %}| calories = {{food['nutritional_info']['energy']}}{% endif %}
{% if 'protein' in food['nutritional_info'] %}| protein = {{food['nutritional_info']['protein']}}{% endif %}
{% if 'carbohydrates' in food['nutritional_info'] %}| carbohydrates = {{food['nutritional_info']['carbohydrates']}}{% endif %}
{% if 'fat' in food['nutritional_info'] %}| fat = {{food['nutritional_info']['fat']}}{% endif %}
{% raw %}}}{% endraw %}

{% raw %}{{external media {% endraw %}
| topic  =  {{food['name']}}
| image1 = [{{food['img_url']}} {{food['name']}} की तस्वीर, फ़ूडब द्वारा प्रदान की गई ]. <ref>{{cite web|url=https://foodb.ca/ |title=FooDB}}</ref>
{% raw %}}}{% endraw %}

===महत्वपूर्ण यौगिक==={% for c in food['important_compounds'] %}
*{{c}}{% endfor %}

===स्वास्थ्य पर पड़ने वाले प्रभाव==={% for c in food['health_effects'] %}
*{{c}}{% endfor %}

===जैविक रास्ते==={% for c in food['pathways'] %}
*{{c}}{% endfor %}"""
```

```python
english_content="""=={{food['name']}}==
{{food['description']}}

{% raw %}{{Infobox {% endraw %} food
| scientific name = {{food['name_scientific']}}
| food group = {{food['food_group']}}
| food subgroup = {{food['food_subgroup']}}
{% if 'energy' in food['nutritional_info'].keys() %}| calories = {{food['nutritional_info']['energy']}}{% endif %}
{% if 'protein' in food['nutritional_info'].keys() %}| protein = {{food['nutritional_info']['protein']}}{% endif %}
{% if 'carbohydrates' in food['nutritional_info'] %}| carbohydrates = {{food['nutritional_info']['carbohydrates']}}{% endif %}
{% if 'fat' in food['nutritional_info'] %}| fat = {{food['nutritional_info']['fat']}}{% endif %}
{% raw %}}}{% endraw %}

===Important Compounds==={% for c in food['important_compounds'] %}
*{{c}}{% endfor %}

===Health Effects==={% for c in food['health_effects'] %}
*{{c}}{% endfor %}

===Blogical Pathways==={% for c in food['pathways'] %}
*{{c}}{% endfor %}"""
```

```python
t = Template(content)
res = t.render(food=data['1'])
print(res)
```

```python
t_e = Template(english_content)
res = t_e.render(food=data['1'])
print(res)
```
